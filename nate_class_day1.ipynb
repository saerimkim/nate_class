```python
hist_df['Year']=hist_df['Year'].str.replace(r"\(.*\)","")
hist_df['20 and 40']=hist_df['20']+hist_df['40']
hist_df.head()
```

```python
def comma_replace(x):
    x=x.str.replace(',','')
    return x
```

```python
hist_df = hist_df.apply(comma_replace)
hist_df.head()
hist_df=hist_df
```

```python
hist_df_subset= hist_df.drop(columns=['20'])
```

```python
type_dict= {
    'Year' : 'datetime64[ns]',
    '20'   : 'int32',
    '40'   : 'int32',
    '60'   : 'int32',
    '80'   : 'int32',
    '95'   : 'int32'
}
```

```python
type_dict
```

```python
hist_df = hist_df.astype(type_dict)
hist_df['add 20 and 40'] = hist_df['20'] + hist_df['40'] #to show it's numeric
hist_df.head()
```

```python
hist_df=hist_df.drop(columns = ['add 20 and 40'])
hist_df.head()
```

```python
hist_df = hist_df.groupby(['Year']).mean()#have to call mean before taking percentage change
hist_df = hist_df.pct_change()
hist_df.tail()
```

```python
hist_df.head()
```

```python
hist_df = hist_df.dropna()

def mult100(x):
    x = x * 100
    return x

# hist_df = hist_df.apply(lambda x: x * 100)
hist_df = hist_df.apply(mult100)
hist_df.tail() #tail of the table
```
